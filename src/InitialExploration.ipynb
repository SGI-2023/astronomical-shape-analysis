```python
import pandas as pd
from sklearn.decomposition import PCA
import plotly.graph_objects as go
import numpy as np
```

```python
df = pd.read_csv("HADES_1_halo_ptcl_z0.ixyz", sep= " ", names=['id', 'x', 'y', 'z'])
df
```

```python
df0 = df.loc[df['id'] == 0]
df0 = df0.drop(['id'], axis=1)
df0
```

```python
P = df0.to_numpy()
pca = PCA(n_components=3)
pca.fit(P)
mean = df0.mean().to_numpy()
cov = pca.get_covariance()
evecs = pca.components_
evals = pca.explained_variance_



print( evals )

evecs[0] = evals[0] * evecs[0]
evecs[1] = evals[1] * evecs[1]
evecs[2] = evals[2] * evecs[2]
```

```python
fig = go.Figure()

fig.add_trace(  
        go.Scatter3d(x = P[:,0], y = P[:,1],  z = P[:,2],
                mode = 'markers',
                marker_size = 1.4,
                name = f'Id = 0') )

for vec in pca.components_:
        print(np.linalg.norm(vec))
        a, b, c = vec + mean
        x_, y_, z_ = mean        
        fig.add_trace(  
        go.Scatter3d(x = [x_, a], y = [y_, b], z = [z_, c],
                        mode= 'lines', 
                        marker_color = 'red',
                        marker_size = 6.0, 
                        name='eigenvector')
        )

fig.update_layout(title= f'Halo Id = {0}')
fig.show()
```
